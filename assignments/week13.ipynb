# 당뇨병 예측

```python
import numpy as np
from sklearn import linear_model
from sklearn import datasets
```

```python
ds = datasets.load_diabetes(scaled=False)
print(ds)
```

```python
x = ds["data"]
y = ds["target"]
print(x[0])
print(y[0])
print(ds["feature_names"])
```

```python
bmi = x[:, 2:3]
bmi
```

```python
reg = linear_model.LinearRegression()
```

```python
reg.fit(bmi, y)
```

```python
reg.predict([[17]])
```

```python
reg.score(bmi, y)
```

```python
import matplotlib.pyplot as plt
plt.scatter(bmi, y)
plt.show()
```
